```python
import requests

def wikipedia_helper(links: list):
    for link in links:
        title = link.split("/")[-1]
        print(title)
        url = "https://en.wikipedia.org/w/api.php"
        params = {
            "action": "query",
            "prop": "pageprops",
            "titles": title,
            "format": "json"
        }

        headers = {
            "User-Agent": "MyBot/1.0 (https://example.com/contact)"  # template header which bypasses 403
        }

        response = requests.get(url, params=params, headers=headers)

        if response.status_code == 200:
            data = response.json()
        else:
            print(f"Failed: {response.status_code}")

        if '-1' in data["query"]["pages"]:
            print(f"error in {link}")

wikipedia_helper(data)
```

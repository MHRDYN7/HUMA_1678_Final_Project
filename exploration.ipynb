```python
import wikipedia

url = "https://en.wikipedia.org/wiki/Han_Kang"

# Split by the standard Wikipedia path and take the last part
title = url.split("/wiki/")[-1]

# # Replace underscores with spaces (optional but cleaner)
# title = title.replace("_", " ")

# Fetch the exact page
page = wikipedia.page(title)
print("Title Found:", page.title)
print("Summary preview:", page.summary[:100], "...")
```
